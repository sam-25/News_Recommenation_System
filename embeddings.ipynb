## using BERT to get the embeddings for the news article

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
for i in range(len(news)):
    news['processed2'][i]=[news['processed2'][i]]
```

```python
news.head()
```

```python
arr=[]
```

```python
for i in range(len(news)):
    arr.append(model.encode(news['processed2'][i]).reshape(-1))
```

```python
news['encoding']=arr
```

```python
news.head()
```

```python
user_rep=[]
```

```python
for i in user['history_list']:
    empty=np.zeros(384)
    for j in i:
        empty+=np.array(dict[j])
    if not len(i):
        empty/=len(i)
    user_rep.append(empty)
```

```python
user['embedding']=user_rep
```
